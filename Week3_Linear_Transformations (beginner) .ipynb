## Setup
We import the basic libraries needed for linear algebra and plotting.

```
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
```

## Part 1: Applying a Linear Transformation
A linear transformation T can always be written as T(x) = A x, where A is a matrix.

```
# Rotation matrix (45 degrees)
theta = np.pi / 4
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

x = np.array([1, 0])  # unit vector
R @ x
```

The output shows where the vector (1,0) moves under rotation.

### Visualizing a Transformation
Blue = original shape, Red = transformed shape.

```
def plot_transform(A):
    square = np.array([[0,1,1,0,0],[0,0,1,1,0]])
    new_square = A @ square

    plt.plot(square[0], square[1], 'b')
    plt.plot(new_square[0], new_square[1], 'r')
    plt.axis('equal')
    plt.grid()
    plt.show()

plot_transform(R)
```

## Part 2: Image and Kernel
- **Image**: all possible outputs of the transformation
- **Kernel**: all vectors that map to zero

```
A = Matrix([[1, 2, -1, 3],
            [2, 4, 0, 8],
            [-1, -2, 3, -1]])
A
```

```
A.rref()
```

Pivot columns of the RREF give a basis for the image.

```
A.nullspace()
```

The vectors above form a basis for the kernel.

## Part 3: Injective and Surjective
- Injective: kernel contains only the zero vector
- Surjective: image fills the whole output space

```
def check_properties(A):
    A = Matrix(A)
    r = A.rank()
    m, n = A.shape
    print('Injective:', r == n)
    print('Surjective:', r == m)

check_properties([[2,1],[5,3]])
```

## Part 4: Matrix Inverses
Only square, full-rank matrices have inverses.

```
A = np.array([[2,1],[5,3]])
A_inv = np.linalg.inv(A)
A_inv
```

```
A @ A_inv
```

The result is the identity matrix, confirming the inverse is correct.
