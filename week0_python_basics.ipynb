## 3. Lists
A **list** holds multiple values in order. In linear algebra, we'll use these to build vectors and matrices.


````python
# A list of numbers
row = [1, 2, 3, 4]
print('List:', row)

# Accessing elements -- indexing starts at 0!
print('First element:', row[0])   # index 0
print('Third element:', row[2])   # index 2
print('Last element: ', row[-1])  # -1 means last

# Length of a list
print('Length:', len(row))

# A list of lists = a matrix row-by-row
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]
print('\nRow 1:', matrix[0])
print('Entry [1][2]:', matrix[1][2])  # row 1, column 2
````

## 5. NumPy Arrays — Vectors and Matrices
In NumPy, both vectors and matrices are **arrays**.  
- A **1D array** is a vector  
- A **2D array** is a matrix  


````python
# Creating a vector
v = np.array([3, -1, 2])
print('Vector v:', v)
print('Shape:', v.shape)    # (3,) means 3 elements

# Creating a matrix
A = np.array([[1, 2],
              [3, 4]])
print('\nMatrix A:')
print(A)
print('Shape:', A.shape)    # (2, 2) means 2 rows, 2 columns
````

## 6. Math with NumPy Arrays
NumPy operations apply **element-by-element** for addition and scalar multiplication —  
exactly how we define vector operations in linear algebra.


````python
u = np.array([1, 2, 3])
v = np.array([4, 0, -1])

print('u =', u)
print('v =', v)
print()
print('u + v    =', u + v)         # vector addition
print('3 * u    =', 3 * u)         # scalar multiplication
print('2u - v   =', 2*u - v)       # linear combination
print('u dot v  =', np.dot(u, v))  # dot product

# Matrix-vector multiplication uses @
A = np.array([[1, 2],
              [3, 4]])
x = np.array([1, -1])
print('\nA @ x =', A @ x)           # @ is matrix multiply
````

## 7. Printing Nicely with f-strings
Throughout the course notebooks you'll see `f'...'` strings.  
These let you embed variables directly inside a printed message using `{}`.


````python
x = 42
name = 'eigenvalue'

# Old way (works but clunky)
print('The', name, 'is', x)

# f-string way (cleaner)
print(f'The {name} is {x}')
print(f'x squared = {x**2}')
print(f'Pi approx = {np.pi:.4f}')  # :.4f means 4 decimal places
````

## 9. SymPy — Exact Math
NumPy is fast but uses **floating point** (small rounding errors are possible).  
SymPy does **exact** arithmetic — fractions stay as fractions, no rounding.

We mainly use SymPy for **RREF** (row reduction), which needs exact results.


````python
from sympy import Matrix, Rational

# NumPy floating point
print('NumPy: 1/3 + 1/3 + 1/3 =', 1/3 + 1/3 + 1/3)  # might not be exactly 1

# SymPy exact
a = Rational(1, 3)   # exact fraction 1/3
print('SymPy: 1/3 + 1/3 + 1/3 =', a + a + a)           # exactly 1

# SymPy matrix
M = Matrix([[1, 2, 3],
            [4, 5, 6]])
print('\nSymPy Matrix:')
print(M)

rref, pivots = M.rref()
print('\nRREF:')
print(rref)
print('Pivot columns:', pivots)
````
