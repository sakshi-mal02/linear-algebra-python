## Part 1: Applying a Linear Transformation (AT1, AT2)


Every linear transformation **T: Rⁿ → Rᵐ** is completely described by a matrix **A**:  
> **T(x) = Ax**  

To apply the transformation, just multiply the matrix by the vector using **`@`**.


```python
# Rotation by 45 degrees
theta = np.pi / 4
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])
print('Rotation matrix (45°):')
print(np.round(R, 4))

x = np.array([1, 0])  # unit vector along x-axis
Tx = R @ x            # apply transformation
print(f'\nT({x}) = {np.round(Tx, 4)}')
print('(Should be [cos45°, sin45°] ≈ [0.7071, 0.7071])')
```

Now let's **visualize** what these transformations do to the unit square:


```python
def transform_and_plot(A, title='Linear Transformation'):
    square = np.array([[0,1,1,0,0],[0,0,1,1,0]])
    pts = np.random.rand(2, 60)
    T_square = A @ square
    T_pts = A @ pts
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    fig.suptitle(title, fontsize=13, fontweight='bold')
    for ax, sq, dots, label in zip(axes, [square,T_square], [pts,T_pts], ['Input','Output']):
        ax.plot(sq[0], sq[1], 'b-', lw=2)
        ax.fill(sq[0], sq[1], alpha=0.15, color='blue')
        ax.plot(dots[0], dots[1], 'r.', alpha=0.5, markersize=4)
        ax.axhline(0,color='k',lw=0.5); ax.axvline(0,color='k',lw=0.5)
        ax.grid(True,alpha=0.3); ax.set_aspect('equal'); ax.set_title(label)
    plt.tight_layout(); plt.show()

transform_and_plot(R, 'Rotation by 45°')

S = np.array([[1.,1.],[0.,1.]])  # shear
transform_and_plot(S, 'Horizontal Shear')

P = np.array([[1.,0.],[0.,0.]])  # projection
transform_and_plot(P, 'Projection onto x-axis')
```

## Part 2: Image and Kernel (AT3)


- **Image** (column space) = all possible outputs of T = span of the columns of A  
- **Kernel** (null space) = all vectors that map to **0**  
- **Rank-Nullity Theorem:** dim(image) + dim(kernel) = n (number of input dimensions)  


```python
A = Matrix([
    [ 1,  2, -1,  3],
    [ 2,  4,  0,  8],
    [-1, -2,  3, -1]
])
print('Matrix A:')
print(A)

rref, pivots = A.rref()
print(f'\nRREF:\n{rref}')
print(f'Pivot columns: {pivots}')

print(f'\nRank (dim of image) = {len(pivots)}')
print('Basis for image (pivot columns of original A):')
for p in pivots:
    print(f'  Column {p+1}: {list(A.col(p).T)}')

null_basis = A.nullspace()
print(f'\nNullity (dim of kernel) = {len(null_basis)}')
print('Basis for kernel:')
for i, v in enumerate(null_basis):
    print(f'  Vector {i+1}: {list(v.T)}')

print(f'\nRank-Nullity: {len(pivots)} + {len(null_basis)} = {len(pivots)+len(null_basis)} == n_cols = {A.cols}: {len(pivots)+len(null_basis)==A.cols}')
```

## Part 3: Injective and Surjective (AT4)


| Property | Condition | Meaning |
|---|---|---|
| **Injective** (one-to-one) | rank = n (pivot in every column) | Kernel = {0} |
| **Surjective** (onto) | rank = m (pivot in every row) | Image = Rᵐ |
| **Bijective** (isomorphism) | both | Matrix is invertible |


```python
def analyze_transformation(A_list, name='T'):
    A = Matrix(A_list)
    m, n = A.shape
    rank = len(A.rref()[1])
    injective  = (rank == n)
    surjective = (rank == m)
    print(f'T: R^{n} -> R^{m},  rank = {rank}')
    print(f'  Injective?  {"YES" if injective  else "NO"}')
    print(f'  Surjective? {"YES" if surjective else "NO"}')
    if injective and surjective: print('  => ISOMORPHISM (invertible matrix)')
    print()

analyze_transformation([[2,1],[5,3]], 'T1 (R²->R²)')
analyze_transformation([[1,2,3],[0,1,-1]], 'T2 (R³->R²)')
analyze_transformation([[1,0],[0,1],[1,1]], 'T3 (R²->R³)')
```

## Part 4: Matrix Inverse (MX2)


The inverse A⁻¹ **undoes** the transformation. Only square, full-rank matrices have inverses.  
Use `np.linalg.inv()` (fast, numerical) or `Matrix.inv()` (exact, SymPy).


```python
A = np.array([[2., 1.],[5., 3.]])
A_inv = np.linalg.inv(A)
print('A^(-1) ='); print(np.round(A_inv, 6))
print('\nA @ A^(-1) (should be identity):'); print(np.round(A @ A_inv, 6))

# Exact inverse with SymPy
A_sym = Matrix([[2,1],[5,3]])
print('\nExact inverse (SymPy):')
print(A_sym.inv())
```
